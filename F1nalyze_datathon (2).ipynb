```python
def convert_to_minutes(time_string):
    if pd.isna(time_string):
        return None
    if isinstance(time_string, float):
        return None  
    
    time_string = str(time_string) 
    
    if ':' in time_string:
        parts = time_string.split(':')
        if len(parts) == 2:  # format is MM:SS.SSS
            minutes = float(parts[0]) + float(parts[1]) / 60
        elif len(parts) == 3:  # format is HH:MM:SS.SSS
            minutes = float(parts[0]) * 60 + float(parts[1]) + float(parts[2]) / 60
        else:
            return None
    else:
        minutes = float(time_string) / 60 
    return minutes

train['time_x'] = train['time_x'].apply(convert_to_minutes)

validation['time_x'] = validation['time_x'].apply(convert_to_minutes)
```
